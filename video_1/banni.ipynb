# merge

```python
df1=df.groupby('Year')[['Grade_1_Dropout','Grade_2_Dropout','Grade_3_Dropout','Grade_4_Dropout']].agg(['mean','min','max'])
df2=df.groupby('Year')[['Grade_5_Dropout','Grade_6_Dropout','Grade_7_Dropout','Grade_8_Dropout']].agg(['mean','min','max'])
```

```python
df1.merge(df2, left_index=True, right_index=True)
```

```python
# 1. Group and Aggregate
df1 = df.groupby('Year')[['Grade_1_Dropout', 'Grade_2_Dropout']].agg(['mean', 'min', 'max'])
df2 = df.groupby('Year')[['Grade_5_Dropout', 'Grade_6_Dropout']].agg(['mean', 'min', 'max'])

# 2. Flatten the MultiIndex Columns
# This joins 'Grade_1_Dropout' and 'mean' into 'Grade_1_Dropout_mean'
df1.columns = ['_'.join(col).strip() for col in df1.columns.values]
df2.columns = ['_'.join(col).strip() for col in df2.columns.values]

# 3. Perform the Merge
# We use reset_index() so 'Year' becomes a regular column we can use with 'on'
final_df = df1.reset_index().merge(
    df2.reset_index(), 
    on='Year', 
    how='inner',       # Only keeps years present in both
    indicator=True     # Adds a column showing where the row came from
)

print(final_df.head())
```
